```python
import pandas as pd
import re
# Carga de la base de datos de rotten Tomatoes-movies-and-critic-reviews-dataset

movies_df = pd.read_csv(
    "/kaggle/input/rotten-tomatoes-movies-and-critic-reviews-dataset/rotten_tomatoes_movies.csv"
)
movies_df.head()
```

```python
# Carga de la base de datos de rotten-tomatoes-movies-and-critic-reviews-dataset
reviews_df = pd.read_csv(
    "/kaggle/input/rotten-tomatoes-movies-and-critic-reviews-dataset/rotten_tomatoes_critic_reviews.csv"
)
reviews_df.head()
```

## Preprocesamiento de rotten_tomatoes_movies.csv.

```python
# Concatena titulo, genero, sinopsis y conseso en una sola columna
movies_df["text"] = (
    movies_df["movie_title"].fillna("") + " " +
    movies_df["genres"].fillna("") + " " +
    movies_df["movie_info"].fillna("") + " " +
    movies_df["critics_consensus"].fillna("")
)
movies_df["text"].head()
```

```python
# Normaliza el texto a minuscilas, elimna caracteres especiales, limpia espacios
def normalize_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
movies_df["normalized_text"] = movies_df["text"].apply(normalize_text)
movies_df["normalized_text"].head()
```

```python
# Divide el texto en palabras individuales de 2 + letras
def tokenize(text):
    return re.findall(r'\b[a-z]{2,}\b', text)
movies_df["tokens"] = movies_df["normalized_text"].apply(tokenize)
movies_df["tokens"].head()
```

```python
# Elimina palabras comunes sin significado basado en el diccionario
STOPWORDS = {
    "the", "is", "in", "and", "to", "of", "a", "for", "on", "that", "it",
    "this", "as", "with", "by", "an", "are", "from", "at", "be", "was",
    "or", "which", "we", "can", "has", "have", "will", "not", "if", "but",
    "about", "there", "their", "they", "you", "your", "our", "my", "me",
    "all", "also", "so", "what", "when", "where", "how", "who", "do", "does",
    "did", "would", "could", "should", "may", "might", "must", "been", "being",
    "am", "i"
}
def remove_stopwords(tokens):
    return [word for word in tokens if word not in STOPWORDS]
movies_df["filtered_tokens"] = movies_df["tokens"].apply(remove_stopwords)
movies_df["filtered_tokens"].head()
```

```python
# Reduce palabras a su raiz
def simple_stem(word):
    word = re.sub(r'(ss|ies|s)$', '', word)
    word = re.sub(r'(ing|ed)$', '', word)
    word = re.sub(r'(er|est)$', '', word)
    word = re.sub(r'ly$', '', word)
    return word

def stem_tokens(tokens):
    return [simple_stem(token) for token in tokens]

movies_df["stemmed_tokens"] = movies_df["filtered_tokens"].apply(stem_tokens)
movies_df["stemmed_tokens"].head()
```

```python
# Funcion que ejecuta todo las funciones 
def preprocess(text):
    normalized = normalize_text(text)
    tokens = tokenize(normalized)
    filtered = remove_stopwords(tokens)
    stemmed = stem_tokens(filtered)
    return stemmed

movies_df["processed_text"] = movies_df["text"].apply(preprocess)
movies_df["processed_text"].head()
```

```python
# Crear el dataset listo para TF-IDF
corpus = movies_df["processed_text"].tolist()
len(corpus)
```

## Preprocesamiento de rotten_tomatoes_critic_reviews.csv

```python
# Solo selecciona columnas necesarias de critic_reviews
reviews_df = reviews_df[[
    "rotten_tomatoes_link",
    "review_type",
    "review_content"
]]
reviews_df.head()
```

```python
# Elimina filas donde la review este vacio
reviews_df = reviews_df.dropna(subset=["review_content"])
reviews_df.shape
```

```python
# Procesa todo el contenido usando la funcion completa ya hecha
reviews_df["processed_text"] = reviews_df["review_content"].apply(preprocess)
```

```python
# Agrupa las reviews de cada pelicula y las concatena en una sola lista de tokens
grouped_reviews = (
    reviews_df
    .groupby("rotten_tomatoes_link")["processed_text"]
    .sum()
    .reset_index()
)
grouped_reviews.head()
```

```python
# Unir movies con las reviews agrupadas
# Left join para mantener todas las películas
combined_df = movies_df.merge(
    grouped_reviews[['rotten_tomatoes_link', 'processed_text']], 
    on='rotten_tomatoes_link', 
    how='left',
    suffixes=('_movie', '_reviews')
)

print(f"Peliculas originales: {len(movies_df)}")
print(f"Peliculas con reviews: {len(grouped_reviews)}")
print(f"Dataset combinado: {len(combined_df)}")
```

```python
# Combina tokens procesados de la pelicula con los de las reviews
def combine_tokens(row):
    movie_tokens = row['processed_text_movie'] if isinstance(row['processed_text_movie'], list) else []
    review_tokens = row['processed_text_reviews'] if isinstance(row['processed_text_reviews'], list) else []
    return movie_tokens + review_tokens

combined_df['all_processed_text'] = combined_df.apply(combine_tokens, axis=1)

# Ver estadísticas
combined_df['total_tokens'] = combined_df['all_processed_text'].apply(len)
print("Estadisticas texto combinado:")
print(combined_df['total_tokens'].describe())
```

```python
# Ejemplo
idx = 0
print("=" * 80)
print("PELÍCULA DE EJEMPLO:")
print("=" * 80)
print(f"\nTU¿ITULO: {combined_df.loc[idx, 'movie_title']}")
print(f"\nTOKENS PELICULA: {len(combined_df.loc[idx, 'processed_text_movie'])}")
print(combined_df.loc[idx, 'processed_text_movie'][:20])

if isinstance(combined_df.loc[idx, 'processed_text_reviews'], list):
    print(f"\nTOKENS REVIEWS: {len(combined_df.loc[idx, 'processed_text_reviews'])}")
    print(combined_df.loc[idx, 'processed_text_reviews'][:20])
else:
    print("\nNo tiene reviews")
```

```python
# Guardar las columnas importantes para usarlo en TF-IF
final_columns = [
    'movie_title', 
    'rotten_tomatoes_link',
    'processed_text_movie',
    'processed_text_reviews', 
    'all_processed_text'
]
optional_columns = ['genres', 'tomatometer_rating', 'audience_rating', 'original_release_date']
for col in optional_columns:
    if col in combined_df.columns:
        final_columns.insert(1, col)

processed_df = combined_df[final_columns].copy()
processed_df.head()
```

## Construcción del Sistema

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Convertir listas de tokens a strings
processed_df['text_for_tfidf'] = processed_df['all_processed_text'].apply(lambda x: ' '.join(x))

# Crear el vectorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer(
    max_features=5000,  # Limitar a las 5000 palabras más importantes
    min_df=2,           # Una palabra debe aparecer en al menos 2 documentos
    max_df=0.8          # Ignorar palabras que aparecen en más del 80% de documentos
)

# Crear la matriz TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_df['text_for_tfidf'])
print(f"Matriz TF-IDF creada:")
print(f"Dimensiones: {tfidf_matrix.shape}")
print(f"Películas: {tfidf_matrix.shape[0]}")
```

## Consulta


```python
# Busca una pelicula relevante para una consulta
# Devuelve un  DataFrame con los resultados más relevantes
def search_movies(query, top_k=10, show_details=True):

    # Preprocesa la consulta 
    query_tokens = preprocess(query)
    query_text = ' '.join(query_tokens)
    # Vectoriza la consulta
    query_vector = tfidf_vectorizer.transform([query_text])
    # Calcula similitud coseno con todas las peliculas
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
    # Obtener indices de peliculas más similares
    top_indices = similarities.argsort()[-top_k:][::-1]
    # Crear DataFrame con resultados
    results = processed_df.iloc[top_indices].copy()
    results['similarity_score'] = similarities[top_indices]
    results['rank'] = range(1, top_k + 1)
    if show_details:
        print(f"CONSULTA: '{query}'")
        print(f"Tokens procesados: {query_tokens}")
        
        for idx, row in results.iterrows():
            print(f"\n{row['rank']}. {row['movie_title']}")
            print(f"   Similitud: {row['similarity_score']:.4f}")
            if 'genres' in row:
                print(f"   Generos: {row['genres']}")
            if 'tomatometer_rating' in row:
                print(f"   Rating críticos: {row['tomatometer_rating']}")
            if 'audience_rating' in row:
                print(f"   Rating audiencia: {row['audience_rating']}")
            print("--------------------")
    
    return results[['rank', 'movie_title', 'similarity_score', 'genres', 
                    'tomatometer_rating', 'audience_rating', 'original_release_date']]
```

```python
# Ejemplo de consulta
results_space = search_movies("Love Strange", top_k=10)
```
